# Comparing the Optimizers

In this notebook we are comparing the performance of a set optimizers using a simple MNIST dataset comparing the performance 

**Optimizers**, combined with the Loss Function, enable Machine Learning to work for our data. The loss function, which is a mathematical way of measuring how well your algorithm models your dataset. If our predictions are completely off, our loss function will result in a higher number. If they’re pretty good, it’ll result in a lower number. 

During the training process, we tune and change the parameters (weights) of our model to try and minimize our loss function, and make predictions as correct and optimized as possible. Most machine learning algorithms use some sort of loss function in the process of optimization, or finding the best weights for your data.

**This method of finding the best set of weights for our Neural Network is done by Optimizers**. 

![alt text](https://)

Optimizers shape and mold our model into its most accurate possible form by futzing with the weights. The loss function is the guide to the terrain, enabling the optimizer to move in the right or wrong direction. It’s impossible to know our model’s correct weights from the start. But with some trial and error based on the loss function, we can get the correct good set of weights for our models and thereby better predictions for our model.

```
import os, time
import numpy as np
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.callbacks import TensorBoard
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping, ReduceLROnPlateau
from keras.optimizers import Adam,SGD,Adagrad,Adadelta,RMSprop

from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix, accuracy_score
```

```
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = np.expand_dims(X_train, 3)
X_test = np.expand_dims(X_test, 3)
```

```
optimizers = [
    'Adadelta',
    'Adagrad',
    'Adam',
    'Adamax',
    'Nadam',
    'RMSprop',
    'SGD'
]
for optimizer in optimizers:
  #Declaring a Simple Neural Network 
    model = Sequential()
    model.add(Conv2D(16, (3,3), activation='tanh', input_shape=X_train[0].shape))
    model.add(MaxPooling2D())
    model.add(Conv2D(16, (3,3), activation='tanh'))
    model.add(MaxPooling2D())
    model.add(Conv2D(16, (3,3), activation='tanh'))
    model.add(MaxPooling2D())
    model.add(Flatten())
    model.add(Dense(10, activation='softmax'))
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    #Plotting the Performance of each Optimizer on a simple MNIST dataset. 
    history = model.fit(X_train, y_train, batch_size=32, epochs=5,callbacks=None, validation_split=0.33)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(optimizer)
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()


    
```

```
X = dataset[:,0:10]
Y = dataset[:,10]
```

```
# Preprocessing of the Housing data
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```
from sklearn.model_selection import train_test_split
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```
optimizers = [
    'Adadelta',
    'Adagrad',
    'Adam',
    'Adamax',
    'Nadam',
    'RMSprop',
    'SGD'
]
for optimizer in optimizers:
  #Declaring a Simple Neural Network 
    model_2 = Sequential([
    Dense(1000, activation='relu', input_shape=(10,)),
    Dense(1000, activation='relu'),
    Dense(1000, activation='relu'),
    Dense(1000, activation='relu'),
    Dense(1, activation='sigmoid'),])
    model_2.compile(optimizer=optimizer,loss='binary_crossentropy', metrics=['accuracy'])
    #Plotting the Performance of each Optimizer on a HousingPrices Data
    hist_2 = model_2.fit(X_train, Y_train,batch_size=32, epochs=5,validation_data=(X_val, Y_val))
    plt.plot(hist_2.history['loss'])
    plt.plot(hist_2.history['val_loss'])
    plt.title(optimizer)
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
```

## Conclusion:

From our simulation over the MNIST Data

The Optimizer presidence if :

**RMSProp > NADAM > ADAM > AdaMax > SGD > Adagrad > Adadelta**

From simulation over the Housing Price Data

**NADAM >RMSProp > AdaMax > ADAM > SGD > Adagrad > Adadelta**

So we can concluded the use of the Optimizers is also based on the type of dataset we use and the regularization we perform on the datasets.
